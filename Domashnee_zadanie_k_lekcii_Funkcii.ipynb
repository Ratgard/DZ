```python
#1. перечень всех документов
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
#2.перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}

def comand_p (x):
    for string in documents:
        if string['number'] == x:
            name = string['name']
            return print('Владелец документа:', name)
        elif string['number'] != x:
            continue
    return print('Документ не найден в базе')

def comand_s (x):
    for key, value in directories.items():
        for element in value:
            if x == element:
                return print('Документ хранится на полке:', key)
    return print('Документ не найден в базе')

def comand_l ():
    for doc_1 in documents:
        string = doc_1
        def comand_l_2 (x):
            for k, v in directories.items():
                for number in v:
                    if number == x:
                        print('№:', string['number'], ', тип:' ,string['type'],',', 'владелец:', string['name'],', полка хранения:' ,k )
        comand_l_2 (string['number'])

def comand_ads (x):
    if x in directories.keys():
        print ('Такая полка уже существует. Текущий перечень полок:', ', '.join(directories.keys()))
    else:
        directories[x]=[]
        print('Полка добавлена. Текущий перечень полок:',', '.join(directories.keys()))

def comand_ds (x):
    if x not in directories.keys():
        print('Такой полки не существует. Текущий перечень полок:',', '.join(directories.keys()))
    if x in directories.keys():
        if directories[x] == []:
            del(directories[x])
            print ('Полка удалена. Текущий перечень полок:',', '.join(directories.keys()))
        elif directories[x] != []:
            print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:',', '.join(directories.keys()))

def enter_comand():
    while True:
        x = input('введите команду \n')
        if x == 'p':
            comand_p(input('введите номер документа \n'))
        elif x == 's':
            comand_s(input('введите номер документа \n'))
        elif x == 'ads':
            comand_ads(input('Введите номер полки \n'))
        elif x == 'ds':
            comand_ds(input('Введите номер полки \n'))
        elif x == 'l':
            comand_l ()
        elif x == 'q':
            print('конец')
            break
enter_comand()
```
